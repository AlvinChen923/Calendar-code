```python
import pandas as pd

# 读取Excel文件
excel_file = 'TKP FY2025 Expenses.xlsx'  # 你的Excel文件路径
df = pd.read_excel(excel_file)
df = df.drop([0, 1, 2])
df = df.iloc[:, :17]
df
```

```python
df = pd.DataFrame(df)

# Grouping by 'Expense Code' and 'Description', and summing the numerical columns
new_df = df.groupby(['Expense Code', 'Description']).sum().drop(columns=['Cost Centre', 'Toyota Store Report page', 'Department Report Line'])

# Specifying the file path for the Excel file
excel_file_path = "FY2025-Expense without Cost Centre.xlsx"

# Writing the DataFrame to an Excel file with date format as "m/dd/yyyy"
new_df.to_excel(excel_file_path)

# Printing confirmation
print("DataFrame has been saved to Excel successfully.")
```
